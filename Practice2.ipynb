```python
import numpy as np

a = np.zeros((8,8))
a[1::2, 0::2] = 1 # нечет i, чет j | начинает с [1] до конца с шагом 2 / начинает с [0] до конца с шагом 2
a[0::2, 1::2] = 1 # чет i, нечет j

print(a)
```

```python
import pandas as pd
import numpy as np

s1 = pd.Series([1, 2])
s2 = pd.Series([4, 5])
res1 = np.linalg.norm(s1 - s2)
res2 = np.sqrt((s2[1] - s1[1])**2 + (s2[0] - s1[0])**2)


print(f"Евклид по функции: {res1}")
print(f"Евклид по рассчету: {res2}")
```
