```python
import numpy as np
x = float(input("input x"))
y = float(input("input y"))

str_input = input("Input: +, -,*,/ or e^(x+y), sin(x+y), cos(x+y), x^y.\nq - for quit")
while(str_input != "q"):
    if str_input == '+':
        print(x+y)
    elif str_input == '-':
        print(x-y)
    elif str_input == '*':
        print(x*y)
    elif str_input == '/':
        print(x/y)
    elif str_input == 'e^(x+y)':
        print(np.exp(x+y))
    elif str_input == 'sin(x+y)':
        print(np.sin(x+y))
    elif str_input == 'cos(x+y)':
        print(np.cos(x+y))
    elif str_input == 'x^y':
        print(x**y)
    str_input = input("Input: +, -,*,/ or e^(x+y), sin(x+y), cos(x+y), x^y.\nq - for quit")
    
```
