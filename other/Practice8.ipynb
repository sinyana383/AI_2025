```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X = np.array([[5,3],
             [10,15],
             [15,12],
             [24,10],
             [30,45],
             [85,70],
             [71,80],
             [60,78],
             [55,52],
             [80,91]])

for k in range(2, 10):
    km = KMeans(n_clusters=k).fit(X)
    pred = km.predict(X)
    score = silhouette_score(X, pred)
    print("Кол-во кластеров", k, " ", score)
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

iris = load_iris()

df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

X = df[['petal width (cm)', 'petal length (cm)']] # по двум признакам
for k in range(2, 10):
    km = KMeans(n_clusters=k).fit(X)
    pred = km.predict(X)
    score = silhouette_score(X, pred)
    print("Кол-во кластеров", k, " ", score)
```

```python
import pandas as pd
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])

X = df[['sepal length (cm)', 'sepal width (cm)']]
U = linkage(X, 'complete')

plt.figure(figsize=(10, 7))
plt.xlabel('Образцы')
plt.ylabel('Удаленность')
dendrogram(U)
plt.show()
```
